Programación orientada a objetos: Gestión de una lista de estudiantes
Objetivo: Escribe un programa que defina dos clases, Estudiante y Clase, con atributos y métodos para manipular una lista de estudiantes. El programa debe:

Definir la clase Estudiante con los atributos nombre, edad y nota_final.

Definir el método __str__ para mostrar la información del estudiante.

Definir la clase Clase con una lista de estudiantes.

Incluir métodos en la clase Clase para agregar, eliminar y mostrar estudiantes.

Crear instancias de las clases y utilizar sus métodos.

Explicación detallada

Definir la clase Estudiante:

La clase Estudiante se define con el método __init__ para inicializar los atributos nombre, edad y nota_final cuando se crea una instancia de la clase.

Método __str__:

Este método devuelve una representación en cadena de la información del estudiante.

Definir la clase Clase:

La clase Clase se define con el método __init__ para inicializar una lista vacía de estudiantes.

Método agregar_estudiante:

Este método añade una instancia de Estudiante a la lista de estudiantes.

Método eliminar_estudiante:

Este método elimina un estudiante de la lista basado en el nombre.

Método mostrar_estudiantes:

Este método recorre la lista de estudiantes y muestra la información de cada uno.

Crear instancias de las clases Estudiante y Clase:

Se crean tres instancias de la clase Estudiante con diferentes valores para los atributos.

Se crea una instancia de la clase Clase y se añaden los estudiantes a la clase utilizando el método agregar_estudiante.

Utilizar los métodos de las instancias de Clase:

Se muestran los estudiantes agregados, se elimina un estudiante, y se muestran los estudiantes restantes.

```python
class Estudiante:
    def __init__(self, nombre, edad, nota_final):
        self.nombre = nombre
        self.edad = edad
        self.nota_final = nota_final

    def __str__(self):
        return f"Nombre: {self.nombre}, Edad: {self.edad}, Nota Final: {self.nota_final}"
    
class Clase:
    def __init__(self):
        self.estudiantes = []

    def agregar_estudiante(self, estudiante):
        self.estudiantes.append(estudiante)

    def eliminar_estudiantes(self,nombre):
        self.estudiantes = [estudiante for estudiante in self.estudiantes if estudiante.nombre != nombre]
    
    def mostrar_estudiantes(self):
        for estudiante in self.estudiantes:
            print(estudiante)

estudiante1=Estudiante("Juan", 20, 8.5)
estudiante2=Estudiante("Maria", 22, 9.0)
estudiante3=Estudiante("Pedro", 21, 7.5)
estudiante4=Estudiante("Juan", 20, 8.5)

clase = Clase()
clase.agregar_estudiante(estudiante1)
clase.agregar_estudiante(estudiante2)
clase.agregar_estudiante(estudiante3)

print("Lista de estudiantes antes despues de agregar:")
clase.mostrar_estudiantes()

clase.eliminar_estudiantes("Juan")
print("Lista de estudiantes después de eliminar a Juan:")
clase.mostrar_estudiantes()

```

Instrucciones para usar el programa:
Copia todo este código en un archivo con extensión .py (por ejemplo: gestion_estudiantes.py)

Ejecútalo con Python (python gestion_estudiantes.py)

Sigue las instrucciones que aparecen en pantalla:

Primero te pedirá cuántos estudiantes quieres ingresar

Luego debes ingresar los datos de cada estudiante (nombre, edad, nota)

Después podrás usar el menú para gestionar la lista

El programa incluye validaciones para:

Nombres no vacíos

Edades válidas (números enteros positivos)

Notas válidas (números entre 0 y 10)

Opciones del menú correctas

Puedes agregar, eliminar y visualizar estudiantes cuantas veces necesites antes de salir del programa.

```python
# SISTEMA DE GESTIÓN DE ESTUDIANTES - CÓDIGO COMPLETO

# Definición de la clase Estudiante para representar a cada estudiante
class Estudiante:
    # Método constructor (__init__) que se ejecuta al crear un nuevo Estudiante
    def __init__(self, nombre, edad, nota_final):
        # Atributo para almacenar el nombre del estudiante (tipo string)
        self.nombre = nombre
        # Atributo para almacenar la edad del estudiante (tipo entero)
        self.edad = edad
        # Atributo para almacenar la nota final (tipo float)
        self.nota_final = nota_final

    # Método especial __str__ que define cómo se muestra el objeto como string
    def __str__(self):
        # Retorna un string formateado con los datos del estudiante
        return f"Nombre: {self.nombre}, Edad: {self.edad}, Nota Final: {self.nota_final}"


# Definición de la clase Clase para gestionar un grupo de estudiantes
class Clase:
    # Método constructor que inicializa la lista de estudiantes
    def __init__(self):
        # Lista vacía que almacenará objetos de tipo Estudiante
        self.estudiantes = []

    # Método para agregar un nuevo estudiante a la lista
    def agregar_estudiante(self, estudiante):
        # Añade el objeto estudiante al final de la lista
        self.estudiantes.append(estudiante)

    # Método para eliminar estudiantes por su nombre
    def eliminar_estudiantes(self, nombre):
        # Crea una nueva lista excluyendo los estudiantes con el nombre indicado
        # (Usa comprensión de listas para filtrar)
        self.estudiantes = [est for est in self.estudiantes if est.nombre != nombre]
    
    # Método para mostrar todos los estudiantes en la lista
    def mostrar_estudiantes(self):
        # Verifica si la lista está vacía
        if not self.estudiantes:
            print("No hay estudiantes registrados.")
        else:
            print("\n--- LISTA DE ESTUDIANTES ---")
            # Itera sobre cada estudiante en la lista
            for estudiante in self.estudiantes:
                # Imprime el estudiante (usa el método __str__ definido)
                print(estudiante)
            print("---------------------------")


# Función para ingresar datos de estudiantes mediante input
def ingresar_estudiantes():
    # Crea una nueva instancia de la clase Clase
    clase = Clase()
    
    # Bucle para asegurar que se ingrese un número válido
    while True:
        try:
            # Solicita al usuario el número de estudiantes a ingresar
            cantidad = int(input("\n¿Cuántos estudiantes desea ingresar? "))
            if cantidad > 0:
                break  # Sale del bucle si el número es válido
            print("Por favor ingrese un número mayor a 0.")
        except ValueError:
            print("Error: Debe ingresar un número entero válido.")
    
    # Bucle para ingresar cada estudiante
    for i in range(cantidad):
        print(f"\nIngresando datos del Estudiante {i+1}/{cantidad}:")
        
        # Bucle para validar el nombre (no puede estar vacío)
        while True:
            nombre = input("Nombre del estudiante: ").strip()
            if nombre:  # Verifica que el string no esté vacío
                break
            print("Error: El nombre no puede estar vacío.")
        
        # Bucle para validar la edad (debe ser un entero positivo)
        while True:
            try:
                edad = int(input("Edad del estudiante: "))
                if edad > 0:
                    break
                print("Error: La edad debe ser mayor a 0.")
            except ValueError:
                print("Error: Ingrese un número entero válido.")
        
        # Bucle para validar la nota (debe ser float entre 0 y 10)
        while True:
            try:
                nota_final = float(input("Nota final (0-10): "))
                if 0 <= nota_final <= 10:
                    break
                print("Error: La nota debe estar entre 0 y 10.")
            except ValueError:
                print("Error: Ingrese un número válido.")
        
        # Crea un nuevo objeto Estudiante con los datos ingresados
        nuevo_estudiante = Estudiante(nombre, edad, nota_final)
        # Agrega el estudiante a la clase
        clase.agregar_estudiante(nuevo_estudiante)
        print(f"Estudiante {nombre} agregado correctamente.")
    
    # Retorna el objeto Clase con todos los estudiantes ingresados
    return clase


# Función principal que maneja el menú de opciones
def menu_principal():
    # Crea una instancia de Clase llamando a la función de ingreso
    mi_clase = ingresar_estudiantes()
    
    # Bucle principal del menú
    while True:
        # Muestra las opciones disponibles
        print("\n=== MENÚ PRINCIPAL ===")
        print("1. Mostrar todos los estudiantes")
        print("2. Agregar nuevo estudiante")
        print("3. Eliminar estudiantes por nombre")
        print("4. Salir del programa")
        
        # Solicita la selección al usuario
        opcion = input("Seleccione una opción (1-4): ")
        
        # Opción 1: Mostrar lista de estudiantes
        if opcion == "1":
            mi_clase.mostrar_estudiantes()
        
        # Opción 2: Agregar nuevo estudiante
        elif opcion == "2":
            print("\nAgregar nuevo estudiante:")
            # Reutiliza el código de ingreso para un solo estudiante
            nombre = input("Nombre del estudiante: ").strip()
            while not nombre:
                print("Error: El nombre no puede estar vacío.")
                nombre = input("Nombre del estudiante: ").strip()
            
            edad = 0
            while edad <= 0:
                try:
                    edad = int(input("Edad del estudiante: "))
                    if edad <= 0:
                        print("Error: La edad debe ser mayor a 0.")
                except ValueError:
                    print("Error: Ingrese un número entero válido.")
            
            nota_final = -1
            while not (0 <= nota_final <= 10):
                try:
                    nota_final = float(input("Nota final (0-10): "))
                    if not (0 <= nota_final <= 10):
                        print("Error: La nota debe estar entre 0 y 10.")
                except ValueError:
                    print("Error: Ingrese un número válido.")
            
            mi_clase.agregar_estudiante(Estudiante(nombre, edad, nota_final))
            print(f"Estudiante {nombre} agregado correctamente.")
        
        # Opción 3: Eliminar estudiantes por nombre
        elif opcion == "3":
            nombre = input("\nIngrese el nombre del estudiante a eliminar: ").strip()
            mi_clase.eliminar_estudiantes(nombre)
            print(f"Todos los estudiantes llamados '{nombre}' han sido eliminados.")
        
        # Opción 4: Salir del programa
        elif opcion == "4":
            print("\nGracias por usar el sistema. ¡Hasta luego!")
            break
        
        # Manejo de opciones inválidas
        else:
            print("Opción no válida. Por favor seleccione 1-4.")


# Punto de entrada principal del programa
if __name__ == "__main__":
    print("=== SISTEMA DE GESTIÓN DE ESTUDIANTES ===")
    print("Por favor ingrese los datos de los estudiantes:")
    # Inicia el programa llamando al menú principal
    menu_principal()
```
