```python
# Diccionario para almacenar los datos de los estudiantes
estudiantes = {}

# Funci贸n para agregar un estudiante y sus calificaciones
def agregar_estudiante():
    nombre = input("Ingrese el nombre del estudiante: ")
    materias = {}
    while True:
        materia = input(f"Ingrese una materia para {nombre} (o 'fin' para terminar): ")
        if materia.lower() == 'fin':
            break
        calificacion = int(input(f"Ingrese la calificaci贸n de {materia}: "))
        materias[materia] = calificacion
    
    estudiantes[nombre] = materias
    print(f"\n隆Datos de {nombre} registrados con 茅xito!\n")

# Funci贸n para consultar calificaciones
def consultar_calificacion():
    nombre = input("Ingrese el nombre del estudiante a consultar: ")
    
    if nombre in estudiantes:
        print(f"\nMaterias disponibles para {nombre}:")
        for materia in estudiantes[nombre]:
            print(f"- {materia}")
        
        materia = input("\nIngrese la materia a consultar: ")
        calificacion = estudiantes[nombre].get(materia, "No disponible")
        print(f"\nLa calificaci贸n de {nombre} en {materia} es: {calificacion}\n")
    else:
        print("\n锔 El estudiante no est谩 registrado.\n")

# Men煤 principal
while True:
    print("1. Agregar nuevo estudiante")
    print("2. Consultar calificaci贸n")
    print("3. Salir")
    
    opcion = input("\nSeleccione una opci贸n (1-3): ")
    
    if opcion == "1":
        agregar_estudiante()
    elif opcion == "2":
        consultar_calificacion()
    elif opcion == "3":
        print("\n隆Hasta luego!")
        break
    else:
        print("\nOpci贸n no v谩lida. Intente nuevamente.\n")
```
