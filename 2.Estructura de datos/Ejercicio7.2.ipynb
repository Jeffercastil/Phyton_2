Objetivo: Escribe un programa que utilice un diccionario para almacenar información sobre varios estudiantes y sus calificaciones en diferentes materias. El programa debe realizar las siguientes operaciones:

Crear un diccionario con información sobre tres estudiantes y sus calificaciones en matemáticas, ciencia y literatura.

Utilizar el método .get para obtener y mostrar la calificación de un estudiante específico en una materia específica.

Mostrar un mensaje personalizado si una calificación no está disponible para una materia específica.

Explicación detallada

Definir el diccionario:

Crea un diccionario llamado estudiantes que contenga la información de tres estudiantes y sus calificaciones en diferentes materias (matemáticas, biología y literatura).

Obtener y mostrar la calificación de un estudiante específico en una materia específica:

Ejemplo 1: Obtiene la calificación de Juan en matemáticas utilizando estudiantes.get(estudiante, {}).get(materia, "No disponible"). Si el estudiante o la materia no existen, devuelve "No disponible".

Ejemplo 2: Obtiene la calificación de Ana en literatura de la misma manera.

Ejemplo 3: Obtiene la calificación de Luis en biología.

Ejemplo 4: Intenta obtener la calificación de Luis en matemáticas, que no está disponible, y muestra un mensaje personalizado.


2. Cómo funciona:

Ingresas el nombre del estudiante

Ingresas el nombre del estudiante

```python
estudiantes = {
    "Juan" :{ " Matematicas" : 85, " Historia" : 93, " Lengua" : 72},
    "Pedro" :{ " Matematicas" : 65, " Historia" : 65, " Lengua" : 81},
    "Maria" :{ " Matematicas" : 95, " Historia" : 85}
}

estudiante = "Juan"
materia = "Matematicas"
calificacion = estudiantes.get(estudiante,{}).get(materia,"No disponible")
print(f"La calificación de {estudiante} en {materia} es: {calificacion}")

estudiante = "Ana"
materia = "literatura"
calificacion = estudiantes.get(estudiante, {}).get(materia, "No disponible")
print(f"La calificación de {estudiante} en {materia} es {calificacion}")

estudiante = "Luis"
materia = "biologia"
calificacion = estudiantes.get(estudiante, {}).get(materia, "No disponible")
print(f"La calificación de {estudiante} en {materia} es {calificacion}")

estudiante = "Luis"
materia = "matemáticas"
calificacion = estudiantes.get(estudiante, {}).get(materia, "No disponible")
print(f"La calificación de {estudiante} en {materia} es {calificacion}")

```

```python
# Diccionario para almacenar los datos de los estudiantes
estudiantes = {}

# Función para agregar un estudiante y sus calificaciones
def agregar_estudiante():
    nombre = input("Ingrese el nombre del estudiante: ")
    materias = {}
    while True:
        materia = input(f"Ingrese una materia para {nombre} (o 'fin' para terminar): ")
        if materia.lower() == 'fin':
            break
        calificacion = int(input(f"Ingrese la calificación de {materia}: "))
        materias[materia] = calificacion
    
    estudiantes[nombre] = materias
    print(f"\n¡Datos de {nombre} registrados con éxito!\n")

# Función para consultar calificaciones
def consultar_calificacion():
    nombre = input("Ingrese el nombre del estudiante a consultar: ")
    
    if nombre in estudiantes:
        print(f"\nMaterias disponibles para {nombre}:")
        for materia in estudiantes[nombre]:
            print(f"- {materia}")
        
        materia = input("\nIngrese la materia a consultar: ")
        calificacion = estudiantes[nombre].get(materia, "No disponible")
        print(f"\nLa calificación de {nombre} en {materia} es: {calificacion}\n")
    else:
        print("\n⚠️ El estudiante no está registrado.\n")

# Menú principal
while True:
    print("1. Agregar nuevo estudiante")
    print("2. Consultar calificación")
    print("3. Salir")
    
    opcion = input("\nSeleccione una opción (1-3): ")
    
    if opcion == "1":
        agregar_estudiante()
    elif opcion == "2":
        consultar_calificacion()
    elif opcion == "3":
        print("\n¡Hasta luego!")
        break
    else:
        print("\nOpción no válida. Intente nuevamente.\n")
```
