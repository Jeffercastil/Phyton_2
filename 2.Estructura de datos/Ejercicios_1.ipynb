Manipulación avanzada de listas.
Objetivo: Escribe un programa que realice las siguientes operaciones en una lista de números:

Agregar un número al final de la lista.

Insertar un número en una posición específica.

Eliminar el primer número de la lista.

Encontrar y eliminar un número específico de la lista.

Ordenar la lista en orden ascendente.

Invertir la lista.

Descripción de la solución: En Python, las listas son estructuras de datos versátiles que permiten realizar diversas operaciones. Tu programa debe realizar varias manipulaciones en una lista, como agregar, insertar, eliminar, encontrar y ordenar elementos. Utiliza la lista inicial: [4, 8, 15, 16, 23, 42].

```python
from IPython.display import display, HTML
display(HTML("<style>pre { white-space: pre !important; }</style>"))

# 1. Ingresar la lista original
while True:
    try:
        cantidad = int(input("¿Cuántos números deseas ingresar para la lista? "))
        if cantidad <= 0:
            print("¡Debe ser un número positivo!")
            continue
        break
    except ValueError:
        print("¡Ingresa un número válido!")

numeros = []
for i in range(cantidad):
    while True:
        try:
            num = int(input(f"Ingresa el número {i+1}: "))
            numeros.append(num)
            break
        except ValueError:
            print("¡Debe ser un número entero!")

print("\nLista original:", numeros)

# 2. Menú interactivo mejorado
def mostrar_menu():
    print("\n" + "="*40)
    print("MENÚ PRINCIPAL".center(40))
    print("="*40)
    print("1. Agregar número al final")
    print("2. Insertar número en posición")
    print("3. Eliminar primer número")
    print("4. Eliminar último número")
    print("5. Eliminar número específico")
    print("6. Ordenar lista (ascendente)")
    print("7. Ordenar lista (descendente)")
    print("8. Invertir lista")
    print("9. Mostrar lista actual")
    print("0. Terminar y mostrar resultado final")
    print("="*40)

while True:
    mostrar_menu()
    
    try:
        opcion = input("\nSelecciona una operación (0-9): ")
        
        if opcion == "0":
            break
            
        elif opcion == "1":
            valor = int(input("¿Qué número deseas agregar al final? "))
            numeros.append(valor)
            print(f"✅ Número {valor} agregado al final")
            
        elif opcion == "2":
            valor = int(input("¿Qué número deseas insertar? "))
            posicion = int(input(f"¿En qué posición? (0-{len(numeros)}) "))
            if 0 <= posicion <= len(numeros):
                numeros.insert(posicion, valor)
                print(f"✅ Número {valor} insertado en posición {posicion}")
            else:
                print("❌ Posición inválida")
                
        elif opcion == "3":
            if numeros:
                eliminado = numeros.pop(0)
                print(f"✅ Se eliminó el primer número: {eliminado}")
            else:
                print("❌ La lista está vacía")
                
        elif opcion == "4":
            if numeros:
                eliminado = numeros.pop()
                print(f"✅ Se eliminó el último número: {eliminado}")
            else:
                print("❌ La lista está vacía")
                
        elif opcion == "5":
            valor = int(input("¿Qué número deseas eliminar? "))
            if valor in numeros:
                numeros.remove(valor)
                print(f"✅ Se eliminó el número {valor}")
            else:
                print(f"❌ El número {valor} no existe en la lista")
                
        elif opcion == "6":
            numeros.sort()
            print("✅ Lista ordenada ascendentemente")
            
        elif opcion == "7":
            numeros.sort(reverse=True)
            print("✅ Lista ordenada descendentemente")
            
        elif opcion == "8":
            numeros.reverse()
            print("✅ Lista invertida")
            
        elif opcion == "9":
            print("\nLista actual:", numeros)
            
        else:
            print("❌ Opción no válida. Intenta nuevamente.")
        
        # Mostrar lista después de cada operación (excepto mostrar)
        if opcion not in ("9", "0"):
            print("Lista actual:", numeros)
            
    except ValueError:
        print("❌ Error: Ingresa un valor numérico válido")
    except Exception as e:
        print(f"❌ Error inesperado: {str(e)}")

# 3. Resultado final
print("\n" + "="*40)
print("RESULTADO FINAL".center(40))
print("="*40)
print("Lista modificada:", numeros)
print("="*40)
```
