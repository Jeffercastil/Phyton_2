Manejo de pares clave-valor.
Objetivo: Escribe un programa que utilice un diccionario para almacenar información sobre un inventario de productos. El programa debe realizar las siguientes operaciones:

Crear un diccionario con información sobre cinco productos y sus cantidades.

Añadir dos nuevos productos al diccionario.

Actualizar la cantidad de dos productos existentes.

Eliminar dos productos del diccionario.

Mostrar todos los productos y sus cantidades, ordenados por nombre del producto.

Explicación detallada

Crear un diccionario con información sobre cinco productos y sus cantidades:

Crea un diccionario inventario con cinco productos: "manzanas", "naranjas", "bananas", "peras", y "uvas", cada uno con una cantidad específica.

Muestra el inventario inicial.

Añadir dos nuevos productos al diccionario:

Añade dos nuevos productos "kiwis" con una cantidad de 7 y "mangos" con una cantidad de 13 al diccionario inventario.

Muestra el inventario después de añadir los nuevos productos.

Actualizar la cantidad de dos productos existentes:

Actualiza la cantidad de "manzanas" a 25 y de "naranjas" a 10 en el diccionario inventario.

Muestra el inventario después de actualizar las cantidades de "manzanas" y "naranjas".

Eliminar dos productos del diccionario:

Elimina los productos "bananas" y "uvas" del diccionario inventario utilizando el método .pop().

Muestra el inventario después de eliminar los productos "bananas" y "uvas".

Mostrar todos los productos y sus cantidades, ordenados por nombre del producto:

Utiliza un bucle for para iterar sobre los elementos del diccionario inventario, ordenados por nombre del producto, y mostrar cada producto y su cantidad.

```python
#Solución que debe escribir el estudiante
inventario = {
    "manzanas": 10,
    "naranjas": 15,
    "bananas": 20,
    "peras": 12,
    "uvas": 18
}
print("Inventario inicial:", inventario)

inventario["Kiwi"] = 5
inventario["Mangos"] = 3
print("Inventario después de agregar Kiwi y Mangos:", inventario)

# Actualizar la cantidad de dos productos existentes
inventario["manzanas"] = 25
inventario["naranjas"] = 10
print("Inventario después de actualizar manzanas y naranjas:", inventario)

inventario.pop("bananas")
inventario.pop("peras")
print("Inventario después de eliminar bananas y peras:", inventario)

print("Productos y cantidades ordenados por nombre del producto:")
for producto in sorted(inventario):
    print(f"{producto}: {inventario[producto]}")
```

Flujo del programa:
Inicialización: Crea el diccionario inventario con frutas predeterminadas.

Menú: Muestra las 4 opciones disponibles.

Input: Espera la selección del usuario.

Acciones:

Opción 1: Muestra el inventario ordenado alfabéticamente.

Opción 2: Permite añadir nuevas frutas o modificar cantidades existentes.

Opción 3: Elimina frutas del inventario (si existen).

Opción 4: Muestra el inventario final y cierra el programa.

Validaciones: Maneja errores si el usuario ingresa datos incorrectos.

Ejemplo de ejecución:
1. Ver inventario
2. Agregar/actualizar fruta
3. Eliminar fruta
4. Salir
Selecciona una opción (1-4): 2
Nombre de la fruta: mango
Cantidad de mango: 7

```python
# Diccionario inicial que almacena las frutas y sus cantidades
inventario = {
    "manzanas": 10,   # 10 manzanas disponibles
    "naranjas": 15,   # 15 naranjas disponibles
    "bananas": 20,    # 20 bananas disponibles
    "peras": 12,      # 12 peras disponibles
    "uvas": 18        # 18 uvas disponibles
}

# Función para mostrar el inventario actual
def mostrar_inventario():
    print("\n--- INVENTARIO ACTUAL ---")  # Encabezado
    # Recorre las frutas ordenadas alfabéticamente con sus cantidades
    for fruta, cantidad in sorted(inventario.items()):
        print(f"- {fruta}: {cantidad} unidades")  # Muestra cada fruta con su cantidad
    print("------------------------\n")  # Línea divisoria

# Bucle principal del programa (se ejecuta indefinidamente hasta que el usuario elija salir)
while True:
    # Menú de opciones
    print("\n1. Ver inventario")          # Opción 1: Mostrar inventario
    print("2. Agregar/actualizar fruta")  # Opción 2: Añadir o modificar fruta
    print("3. Eliminar fruta")            # Opción 3: Quitar fruta del inventario
    print("4. Salir")                     # Opción 4: Terminar el programa
    
    # Pide al usuario que seleccione una opción
    opcion = input("Selecciona una opción (1-4): ")
    
    # Opción 1: Mostrar inventario
    if opcion == "1":
        mostrar_inventario()  # Llama a la función que muestra el inventario
    
    # Opción 2: Agregar o actualizar fruta
    elif opcion == "2":
        fruta = input("Nombre de la fruta: ").lower()  # Pide el nombre de la fruta (en minúsculas)
        try:
            # Pide la cantidad (convertida a número entero)
            cantidad = int(input(f"Cantidad de {fruta}: "))
            inventario[fruta] = cantidad  # Actualiza/agrega la fruta al diccionario
            
            # Mensaje dinámico (dice "actualizó" si ya existía, "agregó" si es nueva)
            print(f"\n✅ Se {'actualizó' if fruta in inventario else 'agregó'} {fruta} ({cantidad} unidades)")
        except:
            # Si el usuario no ingresa un número válido
            print("⚠️ Error: La cantidad debe ser un número entero")
    
    # Opción 3: Eliminar fruta
    elif opcion == "3":
        fruta = input("Nombre de la fruta a eliminar: ").lower()  # Pide la fruta a eliminar
        if fruta in inventario:  # Verifica si la fruta existe
            inventario.pop(fruta)  # Elimina la fruta del diccionario
            print(f"\n❌ {fruta.capitalize()} eliminada del inventario")  # Confirmación
        else:
            # Mensaje si la fruta no existe
            print(f"\n⚠️ {fruta.capitalize()} no existe en el inventario")
    
    # Opción 4: Salir del programa
    elif opcion == "4":
        print("\n¡Hasta luego! Inventario final:")
        mostrar_inventario()  # Muestra el inventario final
        break  # Rompe el bucle while para terminar el programa
    
    # Si el usuario ingresa una opción no válida (no es 1, 2, 3 ni 4)
    else:
        print("⚠️ Opción no válida. Intenta de nuevo")
```
