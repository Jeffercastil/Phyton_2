Parámetros y argumentos de una función.
Objetivo: Escribe un programa que defina una función llamada operacion_basica que realice operaciones aritméticas básicas (suma, resta, multiplicación y división) según el operador proporcionado. Este ejercicio te ayudará a entender cómo definir y usar parámetros y argumentos en funciones de Python. El programa debe:

Definir la función operacion_basica con parámetros.

Usar la función pasando diferentes argumentos.

Mostrar los resultados de las operaciones.

Explicación detallada

Definir la función operacion_basica:

Parámetros: La función operacion_basica toma tres parámetros:

a: El primer número.

b: El segundo número.

operador: Una cadena que indica la operación a realizar ('+', '-', '*', '/').

Cuerpo de la función:

Utiliza estructuras condicionales (if, elif, else) para determinar la operación a realizar según el operador proporcionado.

Si el operador es '+', la función devuelve la suma de a y b.

Si el operador es '-', la función devuelve la resta de a y b.

Si el operador es '*', la función devuelve la multiplicación de a y b.

Si el operador es '/', la función verifica si b es diferente de 0 para evitar la división por cero. Si b es diferente de 0, devuelve la división de a entre b, de lo contrario, devuelve un mensaje de error.

Si el operador no es ninguno de los anteriores, la función devuelve un mensaje indicando que el operador no es válido.

Usar la función pasando diferentes argumentos:

Argumentos: Los valores reales que se pasan a la función cuando se llama. Por ejemplo:

operacion_basica(10, 5, '+') pasa los argumentos 10, 5, y '+' a los parámetros a, b, y operador respectivamente.

Se llama a la función operacion_basica con diferentes operadores y se almacenan los resultados en variables.

Se realizan las siguientes operaciones:

Suma: operacion_basica(10, 5, '+')

Resta: operacion_basica(10, 5, '-')

Multiplicación: operacion_basica(10, 5, '*')

División: operacion_basica(10, 5, '/')

División por cero: operacion_basica(10, 0, '/')

Mostrar los resultados de las operaciones:

Se utilizan las funciones print para mostrar los resultados de las operaciones.

```python
def operacion_basica(a,b,operador):
    if operador == '+':
        return a + b
    elif operador == '-':
        return a - b
    elif operador == '*':
        return a * b
    elif operador == '/':
        if b != 0:
            return a / b
        else:
            return "Error: División por cero"
    else:
        return "Error: Operador no válido"

# Uso de la función con diferentes argumentos
resultado_suma = operacion_basica(10, 5, '+')
resultado_resta = operacion_basica(10, 5, '-')
resultado_multiplicacion = operacion_basica(10, 5, '*')
resultado_division = operacion_basica(10, 5, '/')
resultado_division_cero = operacion_basica(10, 0, '/')

# Mostrar resultados
print(f"Suma: {resultado_suma}")
print(f"Resta: {resultado_resta}")
print(f"Multiplicación: {resultado_multiplicacion}")
print(f"División: {resultado_division}")
print(f"División por cero: {resultado_division_cero}")

```

Explicación:

La función operacion_basica ahora tiene un valor por defecto para el parámetro operador ('+')

El usuario ingresa los dos números con input()

Para el operador, si el usuario no ingresa nada (presiona Enter directamente), se usará '+' por defecto

Se muestra el resultado de la operación con el formato "5 + 3 = 8"

Ejemplo de uso:

text
Ingresa el primer número: 10
Ingresa el segundo número: 5
Ingresa el operador (+, -, *, /) [por defecto +]: *
Resultado de 10.0 * 5.0 = 50.0

```python
def operacion_basica(a, b, operador='+'):
    if operador == '+':
        return a + b
    elif operador == '-':
        return a - b
    elif operador == '*':
        return a * b
    elif operador == '/':
        if b != 0:
            return a / b
        else:
            return "Error: División por cero"
    else:
        return "Error: Operador no válido"

# Ingreso de datos por el usuario
num1 = float(input("Ingresa el primer número: "))
num2 = float(input("Ingresa el segundo número: "))
operador = input("Ingresa el operador (+, -, *, /) [por defecto +]: ") or '+'

# Realizar la operación
resultado = operacion_basica(num1, num2, operador)

# Mostrar el resultado
print(f"Resultado de {num1} {operador} {num2} = {resultado}")
```
