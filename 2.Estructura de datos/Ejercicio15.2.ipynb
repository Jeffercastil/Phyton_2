Variables locales y globales.
Objetivo: Escribe un programa que defina varias funciones para manipular un inventario utilizando una variable global y una variable local dentro de otra función. El programa debe:

Definir una variable global para el inventario.

Definir funciones para agregar, eliminar y mostrar productos en el inventario global.

Definir una función que modifique un inventario local y muestre la diferencia con el inventario global.

Realizar llamadas a las funciones y mostrar los resultados.

Explicación detallada

Definir una variable global para el inventario:

Se define una lista vacía llamada inventario que se utilizará como variable global para almacenar los productos.

Definir funciones para manipular el inventario global:

Función agregar_producto: Añade un producto a la lista global inventario y muestra un mensaje de confirmación.

Función eliminar_producto: Elimina un producto de la lista global inventario si existe y muestra un mensaje de confirmación. Si el producto no existe, muestra un mensaje indicando que no se encontró.

Función mostrar_inventario: Muestra el contenido actual de la lista global inventario.

Definir una función que modifique un inventario local:

Función modificar_inventario: Define una lista vacía inventario_local dentro de la función. Luego, define una función anidada agregar_producto_local que añade un producto a inventario_local y muestra un mensaje de confirmación. Llama a esta función anidada para añadir un producto al inventario local y muestra el contenido de inventario_local.

Realizar llamadas a las funciones y mostrar los resultados:

Llama a las funciones agregar_producto, eliminar_producto, mostrar_inventario y modificar_inventario para demostrar el manejo del inventario global y local. Muestra los resultados después de cada operación.

```python
#solución que debe escribir el estudiante
# Variable global para el inventario
inventario = []

def agregar_producto(producto):
    global inventario
    inventario.append(producto)
    print(f"Producto '{producto}' agregado al inventario.")

def eliminar_producto(producto):
    global inventario
    if producto in inventario:
        inventario.remove(producto)
        print(f"Producto '{producto}' eliminado del inventario.")
    else:
        print(f"Producto '{producto}' no encontrado en el inventario.")

def mostrar_inventario():
    print("Inventario actual:", inventario)

def modificar_inventario():
    inventario_local = []
    
    def agregar_producto_local(producto):
        nonlocal inventario_local
        inventario_local.append(producto)
        print(f"Producto '{producto}' agregado al inventario local.")
    
    agregar_producto_local("Producto_local")
    print("Inventario local dentro de modificar_inventario:", inventario_local)

# Llamadas a las funciones
agregar_producto("Manzanas")
agregar_producto("Naranjas")
mostrar_inventario()

eliminar_producto("Manzanas")
mostrar_inventario()

modificar_inventario()
mostrar_inventario()

print("Inventario global final:", inventario)
```

Explicación de las mejoras:
Sistema interactivo: Ahora el programa tiene un menú que se repite hasta que el usuario elija salir.

Inputs para usuario:

input("Ingrese el nombre del producto...") permite al usuario escribir los productos

strip() limpia los espacios en blanco al inicio/final

Muestra numerada: mostrar_inventario() ahora enumera los productos (1. Manzanas, 2. Naranjas)

Control de errores: Si el inventario está vacío, muestra un mensaje apropiado

Estructura clara: El menú principal organiza todas las funcionalidades

Cómo usar el programa:
Al ejecutarlo, verás un menú con opciones

Para agregar: selecciona 1 y escribe el nombre del producto

Para eliminar: selecciona 2 y escribe el nombre exacto

La opción 4 es solo demostrativa (muestra cómo funciona un inventario local)

Selecciona 5 para salir y ver el inventario final

Flujo del programa:
Inicia → Muestra menú

Usuario selecciona opción

Según la opción, solicita datos adicionales con input()

Ejecuta la operación correspondiente

Vuelve al menú (excepto al salir)

```python
# Lista global que almacenará todos los productos del inventario
inventario = []

# Función para agregar productos al inventario
def agregar_producto(producto):
    global inventario  # Indica que usaremos la variable global 'inventario'
    inventario.append(producto)  # Añade el producto al final de la lista
    print(f"Producto '{producto}' agregado al inventario.")  # Confirmación

# Función para eliminar productos del inventario
def eliminar_producto(producto):
    global inventario  # Accedemos a la variable global
    if producto in inventario:  # Verifica si el producto existe
        inventario.remove(producto)  # Elimina la primera ocurrencia del producto
        print(f"Producto '{producto}' eliminado del inventario.")
    else:
        print(f"Producto '{producto}' no encontrado en el inventario.")  # Mensaje de error

# Función para mostrar el contenido actual del inventario
def mostrar_inventario():
    print("\n--- Inventario Actual ---")
    if not inventario:  # Si la lista está vacía
        print("El inventario está vacío.")
    else:
        for i, producto in enumerate(inventario, 1):  # Enumera los productos
            print(f"{i}. {producto}")

# Función que demuestra el uso de un inventario local (no afecta al global)
def modificar_inventario():
    inventario_local = []  # Variable local solo visible dentro de esta función
    
    def agregar_producto_local(producto):
        nonlocal inventario_local  # Accede a la variable del scope superior
        inventario_local.append(producto)
        print(f"Producto '{producto}' agregado al inventario local.")
    
    # Ejemplo de uso del inventario local
    agregar_producto_local("Producto_local")
    print("Inventario local:", inventario_local)

# Función principal que maneja el menú interactivo
def menu_principal():
    while True:  # Bucle infinito hasta que el usuario elija salir
        print("\n=== MENÚ PRINCIPAL ===")
        print("1. Agregar producto")
        print("2. Eliminar producto")
        print("3. Mostrar inventario")
        print("4. Demo inventario local")
        print("5. Salir")
        
        opcion = input("Seleccione una opción (1-5): ")  # Input para selección
        
        if opcion == "1":
            # Agregar producto: solicitamos el nombre del producto
            producto = input("Ingrese el nombre del producto a agregar: ")
            agregar_producto(producto.strip())  # strip() elimina espacios extras
            
        elif opcion == "2":
            # Eliminar producto: solicitamos el nombre del producto
            producto = input("Ingrese el nombre del producto a eliminar: ")
            eliminar_producto(producto.strip())
            
        elif opcion == "3":
            # Mostrar el inventario actual
            mostrar_inventario()
            
        elif opcion == "4":
            # Demostración del inventario local (no afecta al global)
            print("\n--- Demostración de Inventario Local ---")
            modificar_inventario()
            print("(Esto no afecta el inventario global)")
            
        elif opcion == "5":
            # Salir del programa
            print("\nInventario final guardado:")
            mostrar_inventario()
            print("Saliendo del sistema...")
            break  # Rompe el bucle while
            
        else:
            print("Opción no válida. Por favor ingrese un número del 1 al 5.")

# Ejecución del programa
if __name__ == "__main__":
    print("=== SISTEMA DE GESTIÓN DE INVENTARIO ===")
    menu_principal()  # Inicia el menú interactivo
```
