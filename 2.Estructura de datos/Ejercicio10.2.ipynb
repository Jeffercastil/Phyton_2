Programación orientada a objetos: Creación de clase "Cuenta Bancaria".
Objetivo: Escribe un programa que defina una clase llamada CuentaBancaria con atributos y métodos para manipular una cuenta bancaria. El programa debe:

Definir la clase CuentaBancaria con los atributos titular y saldo.

Incluir un método depositar que permita añadir dinero a la cuenta.

Incluir un método retirar que permita retirar dinero de la cuenta, si hay suficiente saldo.

Incluir un método mostrar_saldo que muestre el saldo actual de la cuenta.

Crear instancias de la clase CuentaBancaria y utilizar sus métodos.



```python
class CuentaBancaria:
    def __init__(self, titular,saldo=0):
        self.titular = titular
        self.saldo = saldo

    def depositar(self, cantidad):
            self.saldo += cantidad
            return f"Se han depositado {cantidad} euros. Nuevo saldo: {self.saldo} euros."
    
    def retirar(self, cantidad):
        if cantidad <= self.saldo:
            self.saldo -= cantidad
            return f"Se han retirado {cantidad} euros. Nuevo saldo: {self.saldo} euros."
        else:
            return "Fondos insuficientes."
    
    def mostrar_saldo(self):
        return f"El saldo actual es: {self.saldo} euros."
    
    # Crear instancias de la clase CuentaBancaria
cuenta1 = CuentaBancaria("María")
cuenta2 = CuentaBancaria("Juan", 500)

# Utilizar los métodos de las instancias de CuentaBancaria
print(cuenta1.depositar(100))
print(cuenta1.retirar(50))
print(cuenta1.retirar(60))
print(cuenta1.mostrar_saldo())

print(cuenta2.depositar(200))
print(cuenta2.retirar(100))
print(cuenta2.retirar(700))
print(cuenta2.mostrar_saldo())
```

```python
# Definición de la clase CuentaBancaria que representa una cuenta bancaria básica
class CuentaBancaria:
    # Método constructor (__init__) que se ejecuta al crear una nueva cuenta
    def __init__(self, titular, saldo=0):
        # Atributo: nombre del titular de la cuenta (string)
        self.titular = titular
        # Atributo: saldo actual (float, con valor por defecto 0)
        self.saldo = saldo

    # Método para depositar dinero en la cuenta
    def depositar(self, cantidad):
        # Suma la cantidad al saldo actual
        self.saldo += cantidad
        # Retorna mensaje de confirmación con el nuevo saldo
        return f"Se han depositado {cantidad} euros. Nuevo saldo: {self.saldo} euros."
    
    # Método para retirar dinero de la cuenta
    def retirar(self, cantidad):
        # Verifica si hay suficiente saldo
        if cantidad <= self.saldo:
            # Resta la cantidad del saldo
            self.saldo -= cantidad
            # Retorna mensaje de confirmación
            return f"Se han retirado {cantidad} euros. Nuevo saldo: {self.saldo} euros."
        else:
            # Mensaje de error si no hay fondos
            return "Fondos insuficientes."
    
    # Método para consultar el saldo actual
    def mostrar_saldo(self):
        # Retorna el saldo formateado como string
        return f"El saldo actual es: {self.saldo} euros."

# Función principal del programa
def main():
    # Lista para almacenar todas las cuentas creadas
    cuentas = []
    
    print("¡Bienvenido al Sistema Bancario!")
    
    # Pregunta cuántas cuentas desea crear
    while True:
        try:
            num_cuentas = int(input("\n¿Cuántas cuentas deseas crear? "))
            if num_cuentas > 0:
                break
            else:
                print("Debe ingresar un número mayor a 0")
        except ValueError:
            print("Por favor ingrese un número válido")
    
    # Creación de cada cuenta
    for i in range(num_cuentas):
        print(f"\nDatos de la cuenta {i+1}:")
        titular = input("Nombre del titular: ").strip()
        
        # Validación para el saldo inicial
        while True:
            try:
                saldo = float(input("Saldo inicial: "))
                if saldo >= 0:
                    break
                else:
                    print("El saldo no puede ser negativo")
            except ValueError:
                print("Ingrese un valor numérico válido")
        
        # Crea la cuenta y la agrega a la lista
        cuentas.append(CuentaBancaria(titular, saldo))
        print(f"Cuenta de {titular} creada con éxito!")
    
    # Menú de operaciones
    while True:
        print("\n--- MENÚ PRINCIPAL ---")
        print("1. Mostrar todas las cuentas")
        print("2. Realizar depósito")
        print("3. Realizar retiro")
        print("4. Consultar saldo")
        print("5. Salir")
        
        opcion = input("Seleccione una opción (1-5): ")
        
        # Opción 1: Mostrar todas las cuentas
        if opcion == "1":
            print("\n--- LISTADO DE CUENTAS ---")
            for i, cuenta in enumerate(cuentas, 1):
                print(f"{i}. {cuenta.titular}")
        
        # Opción 2: Realizar depósito
        elif opcion == "2":
            if not cuentas:
                print("No hay cuentas registradas")
                continue
                
            print("\nSeleccione la cuenta:")
            for i, cuenta in enumerate(cuentas, 1):
                print(f"{i}. {cuenta.titular}")
            
            try:
                seleccion = int(input("Número de cuenta: ")) - 1
                cuenta = cuentas[seleccion]
                cantidad = float(input("Cantidad a depositar: "))
                if cantidad > 0:
                    print(cuenta.depositar(cantidad))
                else:
                    print("La cantidad debe ser positiva")
            except (ValueError, IndexError):
                print("Selección inválida")
        
        # Opción 3: Realizar retiro
        elif opcion == "3":
            # (Similar a la opción 2 pero con retiro)
            if not cuentas:
                print("No hay cuentas registradas")
                continue
                
            print("\nSeleccione la cuenta:")
            for i, cuenta in enumerate(cuentas, 1):
                print(f"{i}. {cuenta.titular}")
            
            try:
                seleccion = int(input("Número de cuenta: ")) - 1
                cuenta = cuentas[seleccion]
                cantidad = float(input("Cantidad a retirar: "))
                if cantidad > 0:
                    print(cuenta.retirar(cantidad))
                else:
                    print("La cantidad debe ser positiva")
            except (ValueError, IndexError):
                print("Selección inválida")
        
        # Opción 4: Consultar saldo
        elif opcion == "4":
            if not cuentas:
                print("No hay cuentas registradas")
                continue
                
            print("\nSeleccione la cuenta:")
            for i, cuenta in enumerate(cuentas, 1):
                print(f"{i}. {cuenta.titular}")
            
            try:
                seleccion = int(input("Número de cuenta: ")) - 1
                print(cuentas[seleccion].mostrar_saldo())
            except (ValueError, IndexError):
                print("Selección inválida")
        
        # Opción 5: Salir del programa
        elif opcion == "5":
            print("\n¡Gracias por usar el Sistema Bancario!")
            break
        
        else:
            print("Opción no válida. Intente de nuevo")

# Punto de entrada del programa
if __name__ == "__main__":
    main()
```
