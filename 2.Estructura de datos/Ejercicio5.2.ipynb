Creación y manipulación de objetos.
Objetivo: Escribe un programa que defina una clase Coche con los atributos marca, modelo, y año, y los métodos acelerar y frenar. Luego, crea una lista de objetos Coche, y realiza las siguientes operaciones:

Acelerar cada coche en la lista.

Frenar cada coche en la lista.

Imprimir la información de cada coche.

Descripción de la solución: En Python, las clases y objetos permiten la encapsulación de datos y funcionalidad relacionada. Tu programa debe definir una clase Coche con los atributos y métodos especificados, crear una lista de instancias de Coche, y realizar las operaciones de aceleración, frenado e impresión de información.

```python
class Coche:
    def __init__(self, marca, modelo , año ):
        self.marca = marca
        self.modelo = modelo
        self.año = año
        self.velocidad = 0

    def acelerar(self):
        self.velocidad += 10
        print(f" {self.marca} {self.modelo} está acelerando a {self.velocidad} km/h")

    def frenar(self):
        self.velocidad -= 10
        if self.velocidad < 0:
            self.velocidad = 0
            print(f" {self.marca} {self.modelo} está frenando a {self.velocidad}  km / h")
    
    def infornacion (self):
        print(f" El coche es de la marca {self.marca} y modelo {self.año} - velocidad actual : {self.velocidad} km/h")


coches =[
    Coche("Toyota", "Corolla", 2020),
    Coche("Honda", "Civic", 2019),
    Coche("Ford", "Mustang", 2021)
]

for coche in coches:
    coche.acelerar()

for coche in coches:
    coche.frenar()

for coche in coches:
    coche.infornacion()

    
        
```

```python
class Coche:
    def __init__(self, marca, modelo, año):
        # Inicializamos los atributos del coche
        self.marca = marca
        self.modelo = modelo
        self.año = año
        self.velocidad = 0  # Todos los coches empiezan detenidos
    
    def acelerar(self):
        # Aumentamos la velocidad en 10 km/h
        self.velocidad += 10
        print(f"{self.marca} {self.modelo} aceleró. Velocidad actual: {self.velocidad} km/h")
    
    def frenar(self):
        # Disminuimos la velocidad en 10 km/h (no puede ser negativa)
        self.velocidad = max(0, self.velocidad - 10)  # Usamos max() para evitar velocidades negativas
        print(f"{self.marca} {self.modelo} frenó. Velocidad actual: {self.velocidad} km/h")
    
    def mostrar_informacion(self):
        # Mostramos toda la información del coche
        print(f"\nInformación del coche:")
        print(f"Marca: {self.marca}")
        print(f"Modelo: {self.modelo}")
        print(f"Año: {self.año}")
        print(f"Velocidad actual: {self.velocidad} km/h")

# Función para que el usuario ingrese los datos de los coches
def ingresar_coches():
    coches = []
    print("\n¡Vamos a crear algunos coches!")
    
    while True:
        try:
            cantidad = int(input("¿Cuántos coches deseas agregar? (1-10): "))
            if 1 <= cantidad <= 10:
                break
            else:
                print("Por favor, ingresa un número entre 1 y 10.")
        except ValueError:
            print("¡Eso no es un número válido!")
    
    for i in range(1, cantidad + 1):
        print(f"\nIngresa los datos del coche {i}:")
        marca = input("Marca (ej. Toyota, Ford): ")
        modelo = input("Modelo (ej. Corolla, Mustang): ")
        año = int(input("Año de fabricación (ej. 2020): "))
        
        coche = Coche(marca, modelo, año)
        coches.append(coche)
        print(f"¡Coche {i} agregado con éxito!")
    
    return coches

# Función principal
def main():
    # El usuario ingresa los coches que desea
    lista_coches = ingresar_coches()
    
    # Operaciones con los coches
    print("\n--- Probando aceleración ---")
    for coche in lista_coches:
        coche.acelerar()
        coche.acelerar()  # Aceleramos dos veces para cada coche
    
    print("\n--- Probando frenado ---")
    for coche in lista_coches:
        coche.frenar()
    
    # Mostramos la información final de todos los coches
    print("\n--- Información final de todos los coches ---")
    for coche in lista_coches:
        coche.mostrar_informacion()

# Ejecutamos el programa
if __name__ == "__main__":
    main()
```
