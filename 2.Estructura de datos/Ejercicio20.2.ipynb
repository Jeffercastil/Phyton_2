Funciones generadoras e iteradoras.
Objetivo: Escribe un programa que demuestre el uso de funciones generadoras e iteradores en Python. Este ejercicio te ayudará a comprender cómo funcionan los generadores para producir una secuencia de valores de manera eficiente y cómo los iteradores permiten recorrer estos valores. El programa debe:

Definir una función generadora que produzca una secuencia de números pares.

Usar un iterador para recorrer la secuencia generada.

Mostrar los números generados en la consola.

Explicación detallada

Definir la función generadora generador_pares:

La función generador_pares es una función generadora que produce números pares hasta un límite dado.

La palabra clave yield se utiliza en lugar de return para devolver un valor y pausar la ejecución de la función. Cuando se llama al generador nuevamente, la ejecución se reanuda justo después de la última yield.

La variable num comienza en 0 y se incrementa en 2 en cada iteración, produciendo así números pares.

Usar el generador para crear un iterador:

Al llamar a la función generador_pares(10), se crea un objeto generador que puede ser iterado para producir números pares desde 0 hasta 10.

Recorrer la secuencia generada usando un iterador:

Se usa un bucle for para iterar sobre el generador. En cada iteración, el generador produce el siguiente número par hasta que alcanza el límite.

```python
def generador_pares(limites):
    num=0
    while num <= limites:
        yield num
        num +=2

pares = generador_pares(10)
for par in pares:
    print(par)
```

Explicación del flujo del programa:
Funciones generadoras:

generador_pares(limite): Produce números pares desde 0 hasta el límite

generador_impares(limite): Produce números impares desde 1 hasta el límite

Ambas usan yield para implementar el patrón generador (eficiente en memoria)

Función de validación:

obtener_numero(mensaje): Garantiza que el usuario ingrese un número entero positivo

Función principal:

Pregunta cuántas secuencias desea generar el usuario

Para cada secuencia:

Solicita el límite máximo

Ofrece elegir entre números pares o impares

Genera y muestra la secuencia correspondiente

Manejo de entradas:

Validación robusta de todos los inputs

Opción por defecto si se ingresa una opción inválida

Mensajes claros para el usuario

Características adicionales:
Flexibilidad para generar múltiples secuencias

Elección entre pares o impares para cada secuencia

Eficiente uso de memoria gracias a los generadores

Interfaz amigable con mensajes descriptivos

Validación de todas las entradas del usuario

Ejemplo de uso:
text
=== GENERADOR DE NÚMEROS PARES O IMPARES ===
¿Cuántas secuencias desea generar? 2

Secuencia 1:
Ingrese el límite máximo: 10

Seleccione el tipo de números:
1. Pares
2. Impares
Elija una opción (1/2): 1

Números pares hasta 10:
0 2 4 6 8 10 

Secuencia 2:
Ingrese el límite máximo: 15

Seleccione el tipo de números:
1. Pares
2. Impares
Elija una opción (1/2): 2

Números impares hasta 15:
1 3 5 7 9 11 13 15 

```python
# Función generadora de números pares
def generador_pares(limite):  # Define función generadora con parámetro 'limite'
    num = 0  # Inicializa el primer número par (0)
    while num <= limite:  # Continúa mientras no se supere el límite
        yield num  # Produce/retorna el número actual (función generadora)
        num += 2  # Incrementa al siguiente número par

# Función generadora de números impares
def generador_impares(limite):  # Define función generadora con parámetro 'limite'
    num = 1  # Inicializa el primer número impar (1)
    while num <= limite:  # Continúa mientras no se supere el límite
        yield num  # Produce/retorna el número actual (función generadora)
        num += 2  # Incrementa al siguiente número impar

# Función para obtener un número válido del usuario
def obtener_numero(mensaje):  # Define función para validar entradas numéricas
    while True:  # Bucle hasta obtener un valor válido
        try:  # Intenta convertir el input
            valor = int(input(mensaje))  # Convierte a entero
            if valor >= 0:  # Verifica que sea positivo
                return valor  # Retorna el valor válido
            else:
                print("Por favor ingrese un número positivo.")  # Mensaje para valores negativos
        except ValueError:  # Captura error si no es número
            print("Por favor ingrese un número válido.")  # Mensaje de error

# Función principal
def main():  # Define la función principal del programa
    print("=== GENERADOR DE NÚMEROS PARES O IMPARES ===")  # Título del programa
    
    # Solicita cantidad de secuencias a generar
    cantidad = obtener_numero("¿Cuántas secuencias desea generar? ")
    
    for i in range(cantidad):  # Bucle para cada secuencia
        print(f"\nSecuencia {i+1}:")  # Muestra número de secuencia actual
        
        # Solicita el límite máximo para la secuencia
        limite = obtener_numero("Ingrese el límite máximo: ")
        
        # Menú de selección par/impar
        print("\nSeleccione el tipo de números:")
        print("1. Pares")
        print("2. Impares")
        opcion = input("Elija una opción (1/2): ")  # Solicita elección
        
        # Genera la secuencia según la opción seleccionada
        if opcion == "1":  # Si eligió pares
            print(f"\nNúmeros pares hasta {limite}:")
            generador = generador_pares(limite)  # Crea generador de pares
        elif opcion == "2":  # Si eligió impares
            print(f"\nNúmeros impares hasta {limite}:")
            generador = generador_impares(limite)  # Crea generador de impares
        else:  # Opción inválida
            print("Opción no válida. Se generarán números pares por defecto.")
            generador = generador_pares(limite)  # Generador por defecto (pares)
        
        # Itera sobre el generador e imprime los números
        for numero in generador:  # Recorre cada número generado
            print(numero, end=" ")  # Imprime números en la misma línea separados por espacio
        print()  # Salto de línea al final

# Punto de entrada del programa
if __name__ == "__main__":  # Verifica si el script se ejecuta directamente
    main()  # Llama a la función principal
```
