```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Grafico de dispersión
plt . scatter(x, y)
#titulo en la grafica
plt . title("Gráfico de dispersión")
plt . xlabel("x")
plt . ylabel("y")
#escala de los ejes
plt .show()
```

```python
x = np.linspace(-1, 5, 400)
```

```python
y = x**2 -4*x + 4
```

```python
plt.Figure(figsize=(8, 4)) # tamaño de la figura
plt.plot(x, y, label ='f (x) = x^2 - 4x + 4', color ='orange') #voy a graficar x y y y el label   y el color rojo
plt.title('Gráfica de una función cuadrática') # titulo de la gráfica
plt.xlabel('x')
plt.ylabel('f (x)')
plt.grid(True) # muestre cuadricula
plt.legend() # muestre la leyenda
plt.show() # muestre el grafico
```

```python
x = [4,6,7]
y = [3,5,2] #marker uniendo con lineas y puntos y se puede utilizar 'o', o 's', '^'
plt.plot(x, y, label= 'Personalización de esta', color =(0.1, 0.2, 0.5), linestyle = '-.',marker='s') ##linestyle = ':' color en formato RGB
plt.legend(loc='upper right') #loc es la ubicación de la leyenda
plt.grid(True)
plt.ylabel('Eje y') #etiqueta de la y
plt.xlabel('Eje x') #etiqueta de la x
plt.title('Gráfica de una función cuadrática') # titulo de la gráfica
plt.show()
```

```python
#USO DE SUBPLOT
x = np.linspace(0,10,100)
y_linear = x
y_exponential = np.exp(x)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(x, y_linear, 'r-')
plt.title('Función lineal')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x, y_exponential, 'b-')
plt.title('Función exponencial')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#graficos uno sobre el otro super puestos
plt.figure(figsize=(8,4))
plt.plot ( x, y_linear, label='linear f(x) = x', color = 'red', linestyle='--')
plt.plot ( x, y_exponential, label='exponential f(x) = exp(x)', color = 'blue' ,linestyle='-')
plt.ylim(0,10)
plt.grid(True)
plt.show()
```

```python
#subplots, varios graficos, en una sola linea

#Crear figura y un arreglo de ejees (2x2)
fig, axs = plt.subplots(2,2 ,figsize=(10,10)) #2filas, 2columnas

#datos
x= np.linspace(0,2+np.pi,100)
funcs =np.sin(x), np.cos(x), np.exp(x),np.sqrt(x)

#titulos para cada subgratica
titles = ['Sin', 'Cos', 'Exp', 'Sqrt']

#llenar cada subgrafica con los datos
for ax, func, title in zip(axs.ravel(), funcs, titles):
    ax.plot(x, func)
    ax.set_title(title)
    ax.set_xlabel('x')
    ax.set_ylabel('f(x)')
```
